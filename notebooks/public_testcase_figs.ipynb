```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
ol_path = '../data/step7/target_STATS.ol.da/MEAN_Fort_Cobb.dat'
daobs_path = '../data/step7/target_STATS.smapDAobs/MEAN_Fort_Cobb.dat'
```

```python
dateparse = lambda x: datetime.strptime(x, '%Y %m %d %H %M')
```

```python
model_output_df = pd.read_csv(ol_path, delim_whitespace=True, header=None,
                    parse_dates={'date': list(range(0,5))},
                   date_parser=dateparse, index_col='date',
                   na_values=-9999.0)
daobs_df = pd.read_csv(daobs_path, delim_whitespace=True, header=None,
                    parse_dates={'date': list(range(0,5))},
                   date_parser=dateparse, index_col='date',
                   na_values=-9999.0)
```

```python
plt.scatter(daobs_df.index, daobs_df[5])
```

```python
output_df = model_output_df[[5, 11]]

output_df.columns = ['OL', 'DA']

output_df
```

```python
output_df.plot()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
ax.scatter(daobs_df.index, daobs_df[5], color='lime', label='DA OBS')
for run in output_df.columns:
    ax.plot(output_df[run], label=run)
ax.legend()
```
